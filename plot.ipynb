```python
path = 'results/gail-service-env|exp_1|diffusion-ql|T-5|ms-offline|k-1|0/episode_rewards.npy'
paths = [
    'results/gail-service-env|exp_1|diffusion-ql|T-5|ms-offline|k-1|0/episode_rewards.npy',
    'results/gail-service-env|exp_1|diffusion-bc|T-5|ms-offline|k-1|0/episode_rewards.npy',
    'episode_rewards_ppo.npy',
    'episode_rewards_a2c.npy',
]
import matplotlib.pyplot as plt
import numpy as np

# data = np.load(path)

# plt.plot(data)
# plt.show()

def smooth_data(data, window_size=5):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

# smooth_data(data)

# plt.plot(smooth_data(data))
# plt.show()

```

```python
def plot_reward_curve(paths):
    diffusion_ql = np.load(paths[0])
    diffusion_bc = np.load(paths[1])
    ppo = np.load(paths[2])
    a2c = np.load(paths[3])
    plt.plot(smooth_data(diffusion_ql))
    plt.plot(smooth_data(diffusion_bc))
    plt.plot(smooth_data(ppo))
    plt.plot(smooth_data(a2c))

    plt.grid()
    plt.xlabel('Episode')
    plt.ylabel('Reward')
    plt.legend(['Diffusion QL', 'Diffusion BC', 'PPO', 'A2C'])
    plt.xlim(0, 100)
    plt.show()

plot_reward_curve(paths)
```
